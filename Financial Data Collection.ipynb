```python
companies = pd.read_csv('C:/Users/aniket/Company_finance/companies_1000.csv')
```

```python
company_arr=companies['Company_Name'].values[:300]
```

```python
companies['Company_Name'].nunique()
```

```python
column_list=['Total Liab', 'Total Stockholder Equity', 'Total Assets',
       'Total Current Liabilities', 'Total Current Assets', 'Long Term Debt',
       'Inventory', 'Net Income', 'Gross Profit', 'Ebit', 'Total Revenue',
       'Cost Of Revenue']
```

```python
company_df = pd.DataFrame(columns=company_arr,index=column_list)

not_working=[]
working=[]
nw=0
for elem in company_arr:
    try:
        comp = yf.Ticker(elem)
        print(" \nFor company",elem,"\n")
        col_2019 = [col for col in comp.balancesheet.columns if str(col).startswith('2018')]
        temp=pd.concat([comp.balancesheet,comp.financials],axis=0)[col_2019]
        temp_filtered = temp.loc[temp.index.intersection(column_list),col_2019]
        company_df[elem] = temp_filtered.squeeze().values
        working.append(elem)
    except:
        print("Not working for company",elem)
        not_working.append(elem)
        nw+=1
        print("List size of not working companies is",nw)
        continue
```

```python
company_df.drop(columns=not_working,inplace=True)
```

```python
cdf=company_df.transpose()
```
